```python
import os
import visualize
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.interpolate import spline
from scipy.signal import medfilt

method_name = ["indep", "maddpg", "clde"]
# env_name = ["multiposrealright_normal_a_squaresafeinit_staticobs_vmaxdismax_avgmode_timecost_norewvpref_highfre-v0",
#             "multiposrealright_normal_v_withvprev_squaresafeinit_staticobs_vmaxdismax_avgmode_timecost_norewvpref_highfre-v0",
#             "multimodrealright_normal_a_squaresafeinit_staticobs_vmaxdismax_avgmode_timecost_norewvpref_highfre-v0",
#             "multimodrealright_normal_v_withvprev_squaresafeinit_staticobs_vmaxdismax_avgmode_timecost_norewvpref_highfre-v0"]

env_name = ["50poly_pad_poly_hierarchical_lidar_v-v0"]
# method_name = ["decppo", "indppo"]
# env_name = ["multiposrealright_normal_v_withvprev_squaresafeinit_staticobs_vmaxdismax_avgmode_timecost_norewvpref_highfre-v0",
#             "multiposrealright_normal_a_squaresafeinit_staticobs_vmaxdismax_avgmode_timecost_norewvpref_highfre-v0"]

log_dir = "./log/"
folderdir = []
for method in method_name:
    methodpath  = os.path.join(log_dir, method+'_log')
#     for env in env_name:
#         envdir =os.path.join(methodpath, env)
#         folderdir.append(envdir)
    folderdir.append(methodpath)
```

```python
# color stuff
#colors = ["salmon pink", "tiffany blue", "light lavender", "peach", "light blue grey", "pastel yellow", "foam green",  ]
colors = ["shamrock green","neon red","cerulean","pinkish purple","mango","mustard yellow", "peach", "light blue grey", "pastel yellow"]
sns.set_palette(sns.xkcd_palette(colors))
sns.palplot(sns.xkcd_palette(colors)) #"leather" "pinkish purple"
```

```python
from scipy.signal import savgol_filter

# convergence
convergence_entry = ['Test index', 'Setting', 'Method', 'Number of frames', 'Accumulated total rewards'] #Setting
#convergence_entry = ['Test index', '_', '__', 'Number of frames', 'Accumulated total rewards'] #Setting
convergence_title = {e:[] for e in convergence_entry}
convergence_data = pd.DataFrame(convergence_title)
#num_iteration =  5000 #100000 #50000 #save-interval 100; frames 1e7
bin_size = 50000
num_iteration = int(30000000/bin_size)
#num_iteration = 100

# randomseed = ["seed1","seed2","seed3"]
# setting = ["indep. PPO P+V","indep. PPO P+A","indep. PPO L+V","indep. PPO L+A","our method P+V","our method P+A","our method L+V","our method L+A"]
# randomseed = ["seed1"] #,"seed2","seed3"]
#setting = ["indep. PPO P+A", "our method P+A"]\

#setting = ["P+V","P+A","L+V","L+A"]
# setting = ["P+A","P+V","L+A","L+V"]
# list_method_name = ["our method","indep. PPO"]

randomseed = ["seed1"]
setting = ["50poly_pad_poly_hierarchical_lidar_v-v0"]
list_method_name = ["Indep. ppo", "Maddpg", "Our method"]

#index = 0 # index of setting
method_id = 0
for method in method_name:
    index = 0 # index of setting
    for env in env_name:
        list_iteration = []
        list_reward = []
        for seed in randomseed:
#             methodpath  = os.path.join(log_dir, method)
#             randomseedpath = os.path.join(methodpath, seed)
#             envdir =os.path.join(randomseedpath, env)
            envdir = os.path.join(log_dir, method+'_log')
            print(envdir)
        
            #folderdir.append(envdir)
            #smooth = 1(1: smooth_reward_curve; 2: medfilt; 3: raw data), bin_size = 100
            tx, ty = visualize.load_data(envdir, 1, bin_size, 3) # load raw data 
            txnew = tx
            tynew = ty
            
            # print(tynew)
            
            #medfilt
            tynew = medfilt(tynew, kernel_size=7)
            #txnew = txnew
            #medfilt for one more time
            
            #using extend instead of append, because tx/ty is a list
            #list_iteration.extend(tx[:num_iteration]) 
            #list_reward.extend(ty[:num_iteration])
            list_iteration.extend(txnew[:num_iteration]) 
            list_reward.extend(tynew[:num_iteration])
            
        print (method, env, seed, num_iteration)
        list_test_index = [index for _ in range(len(list_iteration))]
        list_setting = [setting[index] for _ in range(len(list_iteration))]
        list_method = [list_method_name[method_id] for _ in range(len(list_iteration))]
        frame = [list_test_index, list_setting, list_method, list_iteration, list_reward]
        frame = list(map(list, zip(*frame)))
        index += 1
    
        df = pd.DataFrame(frame, columns=convergence_entry)
        convergence_data = convergence_data.append(df, ignore_index=True)
    
    method_id += 1

#xnew = np.linspace(list_iteration.min(),list_iteration.max(),300) #300 represents number of points to make between T.min and T.max
```

```python
plt.close()
dashline=[(),(1, 3)] #[(),(2, 2)]
plt.figure(1,figsize = (8,8))
plt.rcParams.update({'font.size': 20}) 
ax = plt.gca()
ax.set_xlabel('Number of frames',fontsize=20)
ax.set_ylabel('Accumulated total rewards',fontsize=20,labelpad =-12)
sns.set_context("paper")
sns.lineplot(x="Number of frames", y="Accumulated total rewards", 
             data=convergence_data, ci=None, 
             hue='Method', dashes=dashline) #, markers=True
# hue='Setting', style='Method'  # different line styles are limited to two styles
legend=ax.legend(loc='lower right',title=None,fontsize=13)
legend.get_title().set_fontsize(fontsize = 8)
plt.tick_params(labelsize=20)
#ax.legend(loc='lower right',title=None)
#plt.legend(loc='lower right',title=None)
plt.savefig('./fig/convergence.png')
plt.close()
```
