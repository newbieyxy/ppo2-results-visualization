```python
import os
import visualize
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.interpolate import spline
from scipy.signal import medfilt

log_dir_first = "./newrew/firststage/"
log_dir_second = "./newrew/secondstage/"
log_dir = [log_dir_first, log_dir_second]

methods = ["clde","indep","mappo"]
# methods = ["clde"]
# methods = ["indep"]
# methods = ["mappo"]
# methods = ["clde","mappo"]


seeds = ["seed1","seed2","seed3"]
# seeds = ["seed1"]

folderdir = []
# for log in log_dir:
    # envdir =os.path.join(log_dir, env+test_env)
    # envdir = os.path.join(log, env_name)
    # folderdir.append(envdir)
    # folderdir.append(log)
for method in methods:
    for seed in seeds:
        for log in log_dir:
            folderdir.append(os.path.join(log, method, seed))
```

```python
# color stuff
#colors = ["salmon pink", "tiffany blue", "light lavender", "peach", "light blue grey", "pastel yellow", "foam green",  ]
colors = ["shamrock green","neon red","cerulean","pinkish purple","mango","mustard yellow", "peach", "light blue grey", "pastel yellow"]
sns.set_palette(sns.xkcd_palette(colors))
sns.palplot(sns.xkcd_palette(colors)) #"leather" "pinkish purple"
```

```python
from scipy.signal import savgol_filter

# convergence
convergence_entry = ['Method', 'Number of frames', 'Accumulated total rewards'] #Setting
# convergence_entry = ['Test index', 'Setting', 'Number of frames', 'Accumulated total rewards'] #Setting
# convergence_entry = ['Test index', '_', '__', 'Number of frames', 'Accumulated total rewards'] #Setting
convergence_title = {e:[] for e in convergence_entry}
convergence_data = pd.DataFrame(convergence_title)
bin_size = 50000 # 100
# num_iteration = int(10000000/bin_size)

# setting = ["P+A","P+V","L+A","L+V"]

# setting = ["w4s32","w4s64","w4s128","w8s32","w8s64","w8s128","w16s32","w16s64"]
# setting = ["multimod_withvprev_new5agents_norewvpref_highfre-v0",
#             "multimod_withvprev_new7agents_norewvpref_highfre-v0"]

# setting = 'newrew'
setting = ['clde','indep','mappo']
setting = ['Our method', 'Indep. PPO', 'MAPPO']
# setting = ["clde"]
# setting = ["indep"]
# setting = ["mappo"]
# setting = ["clde","mappo"]

# for env in env_name:
# for log in log_dir:
for idx, method in enumerate(methods):
    for seed in seeds:
        stage_index = 0
        list_iteration = []
        list_reward = []
        for log in log_dir:
            folderdir = os.path.join(log, method, seed, 'log')
            print(folderdir)

            tx, ty = visualize.load_data(folderdir, 1, bin_size) # load raw data 

            if stage_index == 0:
                last_frame = tx[-1] # max(tx)
            else:
                tx = list(np.array(tx) + last_frame + bin_size)
            
            list_iteration.extend(tx) 
            list_reward.extend(ty)

            # list_test_index = [index for _ in range(len(list_iteration))]
            # list_setting = [setting for _ in range(len(list_iteration))]
            list_setting = [setting[idx] for _ in range(len(list_iteration))]
            # frame = [list_test_index, list_setting, list_iteration, list_reward]
            frame = [list_setting, list_iteration, list_reward]
            frame = list(map(list, zip(*frame)))
            stage_index += 1

            df = pd.DataFrame(frame, columns=convergence_entry)
            convergence_data = convergence_data.append(df, ignore_index=True)
            # ----------------------------------------------------------------

#xnew = np.linspace(list_iteration.min(),list_iteration.max(),300) #300 represents number of points to make between T.min and T.max
```
